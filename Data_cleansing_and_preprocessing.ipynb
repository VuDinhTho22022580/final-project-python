```python
import numpy as np
import pandas as pd
```

```python
raw_data = pd.read_csv('./Data/MarinLiekuriva.csv')
raw_data.head()
```

```python
def missing_data(data):
  mdata = {}
  for i in data:
    missing_value = data[i].isna().sum()/len(data)
    mdata.update({i:missing_value})
  return mdata
```

```python
missing_data(raw_data)
```

```python
data = raw_data[['post_id', 'text', 'time', 'image', 'images','video', 'reactors', 'reactions', 'reaction_count', 'comments', 'comments_full', 'shares', 'fetched_time']]
```

```python
data.info()
```

```python
missing_data(data)
```

```python
data.head()
```

```python
for i in range(len(data)):
  if type(data['text'][i]) == float:
    data['text'][i] = ''
```
